# Basic Data Exploration and Statistics
This notebook provides a clean workflow for exploring and analyzing the `Extended_Employee_Performance_and_Productivity_Data.csv` dataset. It covers data loading, inspection, handling missing values, descriptive statistics, and basic visualizations.

## 1. Import Required Libraries
Import essential libraries for data analysis and visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## 2. Load Dataset
Load the dataset from the CSV file.

```python
# Load the dataset
df = pd.read_csv('Extended_Employee_Performance_and_Productivity_Data.csv')
df.head()
```

## 3. View Basic Dataset Information
Display the first few rows, shape, column names, and data types.

```python
# View basic information
df.info()
df.shape
df.columns
df.dtypes
df.head()
```

## 4. Handle Missing Values
Identify and handle missing values in the dataset.

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]

# Example: Drop rows with missing values (can be modified as needed)
df_clean = df.dropna()
df_clean.shape
```

## 5. Descriptive Statistics
Calculate and display summary statistics for numerical columns.

```python
# Descriptive statistics
df.describe().T
```

## 6. Visualize Data Distributions
Create histograms and boxplots for key numerical features.

```python
# Visualize distributions for numerical columns
num_cols = df.select_dtypes(include=np.number).columns

# Histograms
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Boxplots
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col].dropna())
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```

## 7. Correlation Analysis
Compute and visualize the correlation matrix for numerical variables.

```python
# Correlation analysis
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
